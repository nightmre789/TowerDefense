```
import numpy as np
from matplotlib import pyplot as plt

ys = 200 + np.random.randn(200)
x = [x for x in range(len(ys))]

plt.plot(x, ys, '*')
plt.fill_between(x, ys, 200, where=(ys > 100), facecolor='b', alpha=0.9)

plt.title("Sample Visualization")
plt.show()
```
